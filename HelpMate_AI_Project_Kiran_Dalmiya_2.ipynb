#<font color = red> 2. Read, Process, and Chunk the PDF Files
We will be using pdfplumber to read and process the PDF files.

```python
# Define the path of the PDF
pdf_path = '/content/drive/MyDrive/HelpMate AI Project Kiran Dalmiya/Principal-Sample-Life-Insurance-Policy.pdf'
```

```python
def extract_pdf_data(pdf_path):
    """
    Extracts page numbers, section names, and page texts from a PDF document.

    Parameters:
    pdf_path (str): The path to the PDF document.

    Returns:
    pd.DataFrame: A DataFrame containing the page number, section name, and text for each page.

    """
    # Initialize lists to hold the extracted data
    page_numbers = []
    sections = []
    texts = []

    # Open the PDF and extract information
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            # Extract text from the page
            text = page.extract_text()

            # Extract the footer for section name
            footer = page.within_bbox((0, page.height - 70, page.width, page.height)).extract_text()
            section_name = footer.strip() if footer else "Unknown Section"

            # Store the page number, section name, and text
            page_numbers.append(i + 1)
            sections.append(section_name)
            texts.append(text.strip() if text else "No Text Found")

    # Create a DataFrame to store and return the extracted information
    data = {
        'Page Number': page_numbers,
        'Section Name': sections,
        'Page Text': texts
    }
    df = pd.DataFrame(data)


    # Removing the footer/section name from page texts
    # Iterate through the DataFrame and replace section names in each 'Page Text'
    for index, row in df.iterrows():
        df.loc[index, 'Page Text'] = row['Page Text'].replace(row['Section Name'], "")


    # Clean section names
    ## Use a lambda function to remove substrings that start with "GC" and have a total length of up to 7 characters

    df['Section Name'] = df['Section Name'].apply(lambda x: re.sub(r'GC.{0,5}', '', x).strip())



    # Filter out irrelevant rows
    # Removing rows where 'Page Text' is blank and 'Section Name' is "Unknown Section",
    # or 'Page Text' contains the string "This page left blank intentionally"
    df = df[
        df['Page Text'].str.strip().astype(bool) &  # Not blank
        (df['Section Name'] != "Unknown Section") &  # Section name is not "Unknown Section"
        (df['Page Text'].str.strip().str.lower() != "this page left blank intentionally")  # Page does not contain "This page left blank intentionally"
    ]


    #Retain only the rows with a text length of at least 10
    df['Text_Length'] = df['Page Text'].apply(lambda x: len(x.split(' ')))
    df = df.loc[df['Text_Length'] >= 10]


    # Add metadata and filter short texts
    df['Metadata'] = df.apply(lambda x: {'Section Name': x['Section Name'], 'Page Number': x['Page Number']}, axis=1)

    return df

```

Given that the document is 64 pages long and after processing it the maximum text size on a page is no more than 411 words, its a relatively manageable document size. For this scenario, it would still beneficial to use **Semantic Chunking with a Fixed-Size strategy** that aligns with the constraints of the document.

```python
# Function to apply semantic chunking with fixed-size chunks to text

def chunk_text(text, chunk_size):
    words = nltk.word_tokenize(text)
    chunks = [words[i:i + chunk_size] for i in range(0, len(words), chunk_size)]
    return [" ".join(chunk) for chunk in chunks]

# Define the chunk size (e.g., 250 words per chunk)
chunk_size = 250
```

```python
def chunk_text_by_sentences(text, max_words):
    sentences = nltk.sent_tokenize(text)
    chunks = []
    current_chunk = []
    current_chunk_size = 0

    for sentence in sentences:
        sentence_length = len(nltk.word_tokenize(sentence))

        if current_chunk_size + sentence_length <= max_words:
            current_chunk.append(sentence)
            current_chunk_size += sentence_length
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentence]
            current_chunk_size = sentence_length

    # Add the last chunk if any sentences remain
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks
```

```python
# Apply chunking to the text
insurance_pdf['Chunks'] = insurance_pdf['Page Text'].apply(lambda x: chunk_text_by_sentences(x, chunk_size))
```

```python
# Flatten the chunks into a list and create a new dataframe
chunks_list = []
metadata_list = []

for index, row in insurance_pdf.iterrows():
    for chunk in row['Chunks']:
      if len(chunk.strip()) > 0:  # Ensure chunk is not empty   modified *****
         chunks_list.append(chunk)
         metadata_list.append({'Section Name': row['Section Name'], 'Page Number': row['Page Number']})

chunks_df = pd.DataFrame({
    'Chunk Text': chunks_list,
    'Metadata': metadata_list
})

```

# <font color = red> 4. Semantic Search with Cache
In this section, we will perform a semantic search of a query in the collections embeddings to get several top semantically similar results.

we will process three queries here-

1.What is the procedure for filing the claim?

2.Are there any penalties for early withdrawal?

3.What is the grace period for premium payment?

```python
# Read the user Query
query = input("Enter your query: ")
```

```python
# Search the Cache collection first
# Query the collection against the user query and return the top results
cache_results = cache_collection.query(
    query_texts=query,
    n_results=1
)
```

As initially there is no data in cache it will show empty results

```python
results = insurance_collection.query(
    query_texts=query,
    n_results=10
)
```

```python
results
```

```python
# Implementing Cache in Semantic Search

# Set a threshold for cache search
threshold = 0.2

ids = []
documents = []
distances = []
metadatas = []
results_df = pd.DataFrame()


# If the distance is greater than the threshold, then return the results from the main collection.

if cache_results['distances'][0] == [] or cache_results['distances'][0][0] > threshold:
      # Query the collection against the user query and return the top 10 results
      results = insurance_collection.query(
      query_texts=query,
      n_results=10
      )

      # Store the query in cache_collection as document w.r.t to ChromaDB so that it can be embedded and searched against later
      # Store retrieved text, ids, distances and metadatas in cache_collection as metadatas, so that they can be fetched easily if a query indeed matches to a query in cache
      Keys = []
      Values = []

      for key, val in results.items():
        if val is None:
          continue
        for i in range(9):
          Keys.append(str(key)+str(i))
          Values.append(str(val[0][i]))


      cache_collection.add(
          documents= [query],
          ids = [query],  # Or if you want to assign integers as IDs 0,1,2,.., then you can use "len(cache_results['documents'])" as will return the no. of queries currently in the cache and assign the next digit to the new query."
          metadatas = dict(zip(Keys, Values))
      )

      print("Not found in cache. Found in main collection.")

      result_dict = {'Metadatas': results['metadatas'][0], 'Documents': results['documents'][0], 'Distances': results['distances'][0], "IDs":results["ids"][0]}
      results_df = pd.DataFrame.from_dict(result_dict)
      results_df


# If the distance is, however, less than the threshold, you can return the results from cache

elif cache_results['distances'][0][0] <= threshold:
      cache_result_dict = cache_results['metadatas'][0][0]

      # Loop through each inner list and then through the dictionary
      for key, value in cache_result_dict.items():
          if 'ids' in key:
              ids.append(value)
          elif 'documents' in key:
              documents.append(value)
          elif 'distances' in key:
              distances.append(value)
          elif 'metadatas' in key:
              metadatas.append(value)

      print("Found in cache!")

      # Create a DataFrame
      results_df = pd.DataFrame({
        'IDs': ids,
        'Documents': documents,
        'Distances': distances,
        'Metadatas': metadatas
      })
```

#<font color = red>5. Re-Ranking with a Cross Encoder

```python
# Import the CrossEncoder library from sentence_transformers

from sentence_transformers import CrossEncoder, util
```

```python
# Initialise the cross encoder model

cross_encoder = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
```

```python
 # Prepare pairs for re-ranking
cross_inputs = [[query, response] for response in results_df['Documents']]


# Generate the cross_encoder scores for these pairs
cross_rerank_scores = cross_encoder.predict(cross_inputs)
```

```python
# Add the scores to the DataFrame
results_df['Reranked_scores'] = cross_rerank_scores
```

We'll construct a comprehensive prompt that includes:

- The context retrieved from the documents.

```python
# Define the function to generate the response. Provide a comprehensive prompt that passes the user query and the top 3 results to the model

def generate_response(query, results_df):
    """
    Generate a response using GPT-3.5's ChatCompletion based on the user query and retrieved information.
    """

    few_shot_examples = [
        {
            "role": "user",
            "content": """You are a knowledgeable assistant in the insurance domain, skilled at answering user queries about insurance policies and documents.
                          Given a user's question in 'What is the claim process for accidental death?' and search results from a corpus of insurance documents in the dataframe '{top_3_RAG}', your task is to use these search results to provide an informative answer to the query.
                          The dataframe '{top_3_RAG}' includes two key columns: 'documents' contains the actual text from the policy document, and 'metadata' contains the section name and source page.
                          Your task is to:
                          1. Use the relevant information from the 'documents' column to answer the user's query 'What is the claim process for accidental death?'.
                          2. Include relevant/accurate numbers where available.
                          3. Use the 'metadata' column to cite the appropriate policy names and page numbers,ensuring that citations are formatted in bold and italics.
                          4. If you cannot provide a complete answer, offer guidance on where the user might find the relevant information in the cited documents.
                          5. Respond directly to the user with a clear and concise answer, avoiding unnecessary details about internal processes.
                          Your final response should be well-formatted, easy to read, and include citations from the relevant documents."""
        },
        {
            "role": "assistant",
            "content": """The claim process for accidental death involves several steps.
                          First, the claimant must notify the insurance company within 30 days of the incident.
                          The notification should include details such as the date, time, and cause of death.
                          After the initial notification, the claimant must submit a completed claim form along with necessary documents, such as the death certificate and post-mortem report.
                          The insurance company will then evaluate the claim based on the submitted documents. If the claim is approved, the benefit amount specified in the policy will be disbursed to the nominee or legal heir.
                          For more detailed instructions and conditions, refer to PART IV - Section D, Claim Procedures on page 61 and 62 of the policy document.

                          Citations:
                          - Section D - Claim Procedures, PART IV, Pages 61-62."""
        },
        {
            "role": "user",
            "content": """You are a knowledgeable assistant in the insurance domain, skilled at answering user queries about insurance policies and documents.
                          Given a user's question in 'Can I appeal a denied claim?' and search results from a corpus of insurance documents in the dataframe '{top_3_RAG}', your task is to use these search results to provide an informative answer to the query.
                          The dataframe '{top_3_RAG}' includes two key columns: 'documents' contains the actual text from the policy document, and 'metadata' contains the section name and source page.
                          Your task is to:
                          1. Use the relevant information from the 'documents' column to answer the user's query 'Can I appeal a denied claim?'.
                          2. Include relevant/accurate numbers where available.
                          3. Use the 'metadata' column to cite the appropriate policy names and page numbers,ensuring that citations are formatted in bold and italics.
                          4. If you cannot provide a complete answer, offer guidance on where the user might find the relevant information in the cited documents.
                          5. Respond directly to the user with a clear and concise answer, avoiding unnecessary details about internal processes.
                          Your final response should be well-formatted, easy to read, and include citations from the relevant documents."""
        },
         {
            "role": "assistant",
            "content": """Yes, you can appeal a denied claim. If your claim is denied, you have the right to request an appeal within 60 days of receiving the denial notice.
                          The appeal request should include a written statement explaining why you believe the claim was wrongly denied, along with any additional supporting documentation.
                          The appeal will be reviewed by a different team than the one that handled the original claim.
                          The insurance company will provide a final decision on the appeal within 30 days.
                          For detailed steps on the appeal process, refer to PART IV - Section D, Claim Procedures on page 62 of the policy document.

                          Citations:
                          - Section D - Claim Procedures, PART IV, Page 62."""
        }
    ]





    messages = [
                {"role": "system", "content":  "You are a knowledgeable assistant in the insurance domain, capable of efficiently answering user queries about insurance policies and related documents."},
                *few_shot_examples,  # Including few-shot examples to guide the model
                {"role": "user", "content": f"""You are a knowledgeable assistant in the insurance domain, skilled at answering user queries about insurance policies and documents.
                                                Given a user's question in '{query}' and search results from a corpus of insurance documents in the dataframe '{top_3_RAG}', your task is to use these search results to provide an informative answer to the query.
                                                The dataframe '{top_3_RAG}' includes two key columns: 'documents' contains the actual text from the policy document, and 'metadata' contains the section name and source page.
                                                Your task is to:
                                                1.Use the relevant information from the 'documents' column to answer the user's query '{query}'.
                                                2.Include relevant/accurate numbers where available
                                                3.Use the 'metadata' column to cite the appropriate policy names and page numbers,ensuring that citations are formatted in bold and italics.
                                                4.If you cannot provide a complete answer, offer guidance on where the user might find the relevant information in the cited documents.
                                                5.Respond directly to the user with a clear and concise answer, avoiding unnecessary details about internal processes.
                                                Your final response should be well-formatted, easy to read, and include citations from the relevant documents. If the query is not relevant to the document, simply state that the query is irrelevant. Provide your complete answer first, followed by the citations."""
                                                },
              ]

    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages
    )

    return response.choices[0].message.content.split('\n')
```
